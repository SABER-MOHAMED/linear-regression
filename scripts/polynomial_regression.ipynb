## importer les librairies

```python
import pandas as pd # pour manipuler les dodnnées
import matplotlib.pyplot as plt # tous ce qui graphique
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import r2_score # pour evaluer la performance de notre model 
from sklearn.linear_model import LinearRegression
```

### Importation de dataset

```python
dataset = pd.read_csv('../data/temperature_data.csv')
X = dataset.iloc[:, 0:1].values  # toutes les lignes la colonne des dates
y = dataset.iloc[:, 1].values # toutes les lignes la colonne des temperature 
```

Reepresenter le dataset pour mieux comprendre la distribution des données, sous forme d'une nuage

```python
plt.scatter(X, y, color = 'red')
```

ici on ne peut pas utiliser le modele de RL, car il y a pas une relation lineaire entre les variables explicatives et la variable cible (T)

### creer et entraine un modele de regression polynomiale

```python
poly_reg = PolynomialFeatures(degree = 10) # on peut changer le degree pour voir l'impact
X_poly = poly_reg.fit_transform(X) # on transforme notre variable X en X_poly, pour avoir les vlaeurs de X, X^2, X^3
regressor_poly = LinearRegression()
regressor_poly.fit(X_poly, y) # on utilise le model de regression lineaire pour predire les valeurs, on lui donne X_poly et y
y_poly_pred = regressor_poly.predict(X_poly) # on predit les valeurs de y en fonction de X_poly
```

### visualiser le resultat obtenue

```python
plt.scatter(X, y, color = 'red') # on affiche les points de notre dataset
plt.plot(X, y_poly_pred, color = 'blue') # on affiche la courbe de regression
plt.title('Polynomial Regression')
```

### Calculer la performance du modele

```python
r2 = r2_score(y, y_poly_pred) # on evalue la performance de notre model
r2
```

88% est un taux de fiabilité et de performance assez bon pour notre modèle
